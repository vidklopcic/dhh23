```python
import pandas as pd

speeches_subset = pd.read_feather('artefacts/bojan/gb_commons_embeedings_subset.feather')
embeddings_subset = [embeddings[i] for i in speeches_subset.ID]
speeches_subset
```

```python
import numpy as np
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
sns.set(style='white', context='notebook', rc={'figure.figsize': (14, 10)})
```

```python
from umap import UMAP

reducer = UMAP(n_neighbors=100,
               min_dist=1,
               n_components=2,
               metric='cosine')
```

```python
embedding = reducer.fit_transform(embeddings_subset)
```

```python
subset = speeches_subset.Speaker_party_name
speaker_parties = subset.unique()
color_map = {v: i if i < 2 else 2 for i, v in enumerate(speaker_parties)}
print(color_map)
color_map = [color_map[x] for x in subset]
```

```python
# plt.clf()
sentences = [s for s in speeches_subset.speech]
c = [sns.color_palette("husl", 3)[x] for x in color_map]
plt.scatter(
    embedding[:, 0],
    embedding[:, 1],
    c=c,
    s=0.3,
    marker='.',
    edgecolors='none',
)
plt.gca().set_aspect('equal', 'datalim')
plt.title('UMAP projection of the GB Commons dataset', fontsize=24);
plt.savefig('artefacts/bojan/umap.pdf')
with open('artefacts/bojan/umap.pkl', 'wb') as f:
    pickle.dump([
        embedding,
        c,
        sentences
    ], f)
# plt.show()
```
